# Titanic Dataset - Exploratory Data Analysis (EDA)

In this notebook, we summarize the exploratory analysis of the Titanic dataset.
The goal is to understand the distribution of features and how they relate to survival.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/titanic_clean.csv")
df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Age Distribution")
plt.show()

sns.countplot(x="Sex", data=df)
plt.title("Counts of Sex")
plt.show()
```

```python
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Key Findings
- Women had a much higher survival rate than men.
- Passengers in higher classes (1st) survived more often than those in 3rd class.
- Younger passengers (Children) had better survival chances.
- Family size had a mixed effect: being alone was risky, but very large families also had low survival.
- Fare was positively associated with survival (wealthier passengers survived more).
